```python
import matplotlib.pyplot as plt #繪圖函式庫

import numpy as np #矩陣函式庫

from sklearn import datasets, linear_model



diabetes = datasets.load_diabetes()




diabetes_X = diabetes.data[:, np.newaxis, 2]



diabetes_X_train = diabetes_X[:-20]

diabetes_X_test = diabetes_X[-20:]



diabetes_y_train = diabetes.target[:-20]

diabetes_y_test = diabetes.target[-20:]
```

```python
regr = linear_model.LinearRegression()# 建立線性迴
regr.fit(diabetes_X_train,diabetes_y_train)#訓練
#係數
print('Coefficients: \n', regr.coef_)
#均方誤差
print("Mean squared error: %.2f"
    %np.mean((regr.predict(diabetes_X_test)- diabetes_y_test)**2))
#顯示方差分數:1 是完美預測
print('Variance score: %.2f' % regr.score(diabetes_X_test, diabetes_y_test))
#繪品
plt.scatter(diabetes_X_test,diabetes_y_test,color='black')#出試黑點
plt.plot(diabetes_X_test,regr.predict(diabetes_X_test), color='blue',
linewidth=3) # 盡出測試預測的線性迴歸答案
plt.xticks(())
plt.yticks(())
plt.show()#顯示
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn import metrics
#取得鳶尾花的資料
iris = datasets.load_iris()
# 拆分資料80:20 訓練與測試用 2 類標笼答案
iris_X_train, iris_x_test, iris_y_train, iris_y_test= train_test_split(iris.data,iris.target,test_size=0.2)

#KMeans

kmeans = KMeans (n_clusters = 2)
kmeans_fit =kmeans.fit (iris_X_train) #SABAR
print("實際", iris_y_train) #
print("預測", kmeans_fit.labels_) # 顯示預測答案
#調整標簽的數字
iris_y_train [iris_y_train==1]=11
iris_y_train [iris_y_train==0] =1
iris_y_train [iris_y_train==11]=0
print("調整", iris_y_train) #
score = metrics.accuracy_score(iris_y_train, kmeans.predict(iris_X_train))
print('準確率: {0: f}'. format (score)) #
```

```python
##K-means應用：信用卡客戶分群
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import calinski_harabasz_score

df = pd.read_csv('customer.csv')
dict1 = {'男':1, '女':2}
df['性別'].replace(dict1, inplace=True)
df
```

```python
km = KMeans(n_clusters=3)
km.fit(df)
km.labels_
```

```python
df['類別'] = km.labels_
df.head()
```

```python
df2 = df[df['類別']==0]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別']==1]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別']==2]
df3 = df2.iloc[0:30, :]
df3
```

```python
for n in range(2,15): 
  km = KMeans(n_clusters=n)
  km.fit(df)
  metric = calinski_harabasz_score(df, km.labels_) 
  print('群組數量：{}，評分：{}'.format(n, metric))
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('customer.csv')
dict1 = {'男':1, '女':2}
df['性別'].replace(dict1, inplace=True)
df
```

```python
scaler = StandardScaler()
scaler.fit(df)
dfScaled = scaler.transform(df)
dbscan = DBSCAN(eps=0.82, min_samples=5)
dbs = dbscan.fit(dfScaled)
dbs.labels_
```

```python
df['類別'] = dbs.labels_
df.head()
```

```python
df2 = df[df['類別']==0]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別']==1]
df3 = df2.iloc[0:30, :]
df3
```

```python
df2 = df[df['類別']==-1]
df3 = df2.iloc[0:30, :]
df3
```
